```python
import random
from tqdm.notebook import tqdm
import numpy as np

options = ['heads', 'tails']
prizes = []

for n in tqdm(range(0, 100000, 1000)):
    history = []
    for i in range(n):
        prize, current_value = 0, 2
        while True:
            result = random.choice(options)
            if result == 'heads':
                prize += current_value
                current_value *= 2
            else:
                break
        history.append(prize)

    prizes.append(np.array(history).mean())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('white')
sns.lineplot(x=list(range(0, 100000, 1000)), y=prizes)
#plt.ylim([0, 100])
plt.show()
```
