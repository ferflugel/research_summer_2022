```python
import pandas as pd

# read file for reviews
reviews = pd.read_csv('csv_data/yelp_academic_dataset_review.csv')

# group by businesses and sort by most reviews
businesses = reviews.groupby('business_id').count().sort_values(by='text', ascending=False)

businesses
```

```python
# select the 1000 most active businesses
businesses_ids = businesses.reset_index()['business_id'].tolist()[:1000]

# select only the reviews from these businesses
reviews = reviews[reviews['business_id'].isin(businesses_ids)]

reviews
```

```python
from tqdm.notebook import tqdm

business_ratings, business_count  = [], []

# create a list tracking the mean rating of businesses over time
for business in tqdm(businesses_ids):
    business_history, business_n = [], []
    user_reviews = reviews[reviews['business_id'] == business].sort_values(by='date')
    for n in range(100, user_reviews.shape[0], 100):
        first_n_reviews = user_reviews.iloc[:n, :]
        business_history.append(first_n_reviews['stars'].mean())
        business_n.append(n - 100)
    business_ratings.append(business_history)
    business_count.append(business_n)
```

```python
import numpy as np

# normalize user ratings so that they always start at 0
normalized_business_ratings = []
for business_rating in business_ratings:
    normalized_business_ratings.append(np.array(-1 * business_rating[0] + np.array(business_rating)))

# normalize user n so that we have all lines of same length
normalized_business_count = []
for business_n in business_count:
    normalized_business_count.append(np.array(business_n) / business_n[-1])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('white')

# create a setup for the mean path
mean_path = np.zeros(100)
mean_path_x = np.linspace(0, 1, 100)

for ratings, count in zip(normalized_business_ratings, normalized_business_count):
    sns.lineplot(x=count, y=ratings, color=[0.6, 0.6, 1.0], alpha=0.03)
    mean_path += np.interp(mean_path_x, count, ratings)

mean_path /= len(normalized_business_ratings)

plt.axhline(ls='--', alpha=0.5, lw=1, color='k')
sns.lineplot(x=mean_path_x, y=mean_path, color='k', alpha=1, label='Mean Path')
plt.legend(loc=2, frameon=False)

plt.ylabel("Business' Change in Mean Rating")
plt.xlabel('Time')

plt.savefig('figures/business_improvement.png', dpi=250)
plt.show()
```
