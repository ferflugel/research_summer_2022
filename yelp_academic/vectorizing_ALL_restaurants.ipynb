```python
import pandas as pd

# import dataset for businesses
restaurants = pd.read_csv('csv_data/yelp_academic_dataset_business.csv')

# create a column for the price range
restaurants.dropna(subset=['attributes'])
restaurants = restaurants[restaurants['categories'].str.contains('Restaurants', na=False)]

restaurants
```

```python
import pickle
from yelp_functions import get_processed_inputs

# import reviews and the vectorizer
reviews = pd.read_csv('csv_data/yelp_academic_dataset_review.csv')
vectorizer = pickle.load(open('models/restaurants_vectorizer.sav', 'rb'))

# zip the id's and prices together
restaurant_ids = list(reviews.groupby('business_id').mean().index)

# check list of ids
restaurant_ids
```

```python
import numpy as np
from yelp_functions import sum_to_one
from tqdm.notebook import tqdm

# create empty list for restaurant vectors
restaurants_vectors = []

# Important to change [:50000], [50 000:100 000], [100 000:]
restaurant_ids = restaurant_ids[100000:]
path = 'processed_data/all_restaurant_vectors_pt3'

# create vectors for restaurants
for restaurant_id in tqdm(restaurant_ids):

    # set up the initial conditions
    temp_df = reviews[reviews['business_id'] == restaurant_id]
    bow_inputs = get_processed_inputs(temp_df[['text']])

    # get individual BoW representations
    bow_inputs = sum_to_one(vectorizer.transform(bow_inputs).toarray())

    # add up the individual BoW representations
    r_vector = np.sum(bow_inputs, axis=0)
    restaurants_vectors.append(r_vector)

# check results for restaurants vectors
len(restaurants_vectors), restaurants_vectors[0].shape[0], restaurants_vectors[0].sum()
```

```python
ids = list(reviews.groupby('business_id').mean().index)

normalized_restaurants = sum_to_one(restaurants_vectors)
data_to_save = [ids, normalized_restaurants]

with open(path, "wb") as pickle_file:
    pickle.dump(data_to_save, pickle_file)
```
