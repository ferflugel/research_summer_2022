```python
import random

random.seed(0)

raw_data = [[[(1 + j) * random.random() for k in range(10)] for j in range(3)] for i in range(4)]

# Create a column with all the data
processed_data = [raw_data[i][j][k] for i in range(4) for j in range(3) for k in range(10)]
metric = [j for i in range(4) for j in range(3) for k in range(10)]
latent = [i for i in range(4) for j in range(3) for k in range(10)]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('white')
sns.stripplot(x=latent, y=processed_data, hue=metric, jitter=0.02, dodge=True, palette='flare')
plt.savefig('figures/example_strip_plot.png', dpi=200)
plt.show()
```
