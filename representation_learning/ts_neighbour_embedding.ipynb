```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.manifold import TSNE

# import the breast cancer dataset
diagnoses = pd.read_csv('datasets/breast_cancer.csv')

# select columns for training and testing
X = diagnoses[diagnoses.columns[2:]]
y = diagnoses['diagnosis']

# create the embeddings for the data
tsne = TSNE(n_components=2, init='random', random_state=0)
X = tsne.fit_transform(X)

# split data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
X_train
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

# train a new classifier with LR
logistic_regression = LogisticRegression(random_state=0).fit(X_train, y_train)
print('Train accuracy:', logistic_regression.score(X_train, y_train))
print('Test accuracy:', logistic_regression.score(X_test, y_test))

# train a new classifier using SVM
svm_classifier = SVC().fit(X_train, y_train)
print('Train accuracy:', svm_classifier.score(X_train, y_train))
print('Test accuracy:', svm_classifier.score(X_test, y_test))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('white')
sns.set_palette('pastel', 2)

sns.scatterplot(x=X.T[0], y=X.T[1], hue=y, alpha=0.5)
plt.title('tSNE')
plt.legend(frameon=False)

plt.savefig('figures/tsne.png', dpi=250)
plt.show()
```
