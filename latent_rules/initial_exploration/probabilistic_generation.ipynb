```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(0)

# Generate random hidden factors
hidden_factors = np.random.normal(0, 1, size=(2, 1000))

# Crete a function phi(h)
non_linear_factors = np.tanh(hidden_factors)

# Plot distribution of hidden factors
sns.set_style('white')
sns.scatterplot(x=non_linear_factors[0], y=non_linear_factors[1])
plt.xlabel('tanh($h_0$)'); plt.ylabel('tanh($h_1$)')
plt.show()
```

```python
x = np.random.normal(non_linear_factors[0] + 1, abs(non_linear_factors[0] - 2))

# Plot distribution of hidden factors
sns.set_style('white')
sns.scatterplot(x=hidden_factors[0], y=x)
plt.xlabel('h_0'); plt.ylabel('x')
plt.show()
```
